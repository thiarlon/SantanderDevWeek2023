```python
import pandas as pd
df = pd.read_csv('sdw2023.csv')
id=df['ID'].tolist()
nome=df['NOME'].tolist()
idade=df['IDADE'].tolist()
estado=df['ESTADO'].tolist()
cartao_credito=df['CARTAO_DE_CREDITO'].tolist()
seguro_de_vida=df['SEGURO_DE_VIDA'].tolist()
investimento=df['INVESTIMENTO'].tolist()
outros_servicos=df['OUTROS_SERVICOS'].tolist()



```

```python
lista_geral = pd.DataFrame((zip(id, nome,idade,estado,cartao_credito,seguro_de_vida,investimento,outros_servicos)), columns = ['ID', 'NOME','IDADE','ESTADO','CARTÃO DE CRÈDITO','SEGURO DE VIDA','INVESTIMENTO','OUTROS SERVIÇOS'])
print("Relação dos clientes\n",lista_geral)

lista_cliente_cartao_credito = pd.DataFrame((zip(id, nome,idade,estado,cartao_credito,)), columns = ['ID', 'NOME','IDADE','ESTADO','CARTÃO DE CRÈDITO'])
print("Status de clientes com Cartão de Crédito\n",lista_cliente_cartao_credito)

lista_seguro_vida = pd.DataFrame((zip(id, nome,idade,estado,seguro_de_vida)), columns = ['ID', 'NOME','IDADE','ESTADO','SEGURO DE VIDA',])
print("Status de clientes com Seguro de vida\n",lista_seguro_vida)

lista_investimento = pd.DataFrame((zip(id, nome,idade,estado,investimento)), columns = ['ID', 'NOME','IDADE','ESTADO','INVESTIMENTO'])
print("Status de clientes com Investimento\n",lista_investimento)

lista_outros_servicos = pd.DataFrame((zip(id, nome,idade,estado,outros_servicos)), columns = ['ID', 'NOME','IDADE','ESTADO','OUTROS SERVIÇOS'])
print("Status de clientes com outros serviços\n",lista_outros_servicos)


```

```python
lista_geral.to_csv("Relação dos clientes.csv",sep= ",")
lista_cliente_cartao_credito.to_csv("Status de clientes com Cartão de Crédito.csv",sep= ",")
lista_seguro_vida.to_csv("Status de clientes com Seguro de vida.csv",sep= ",")
lista_investimento.to_csv("Status de clientes com Investimento.csv",sep= ",")
lista_outros_servicos.to_csv("Status de clientes com outros serviços.csv",sep= ",")
```
